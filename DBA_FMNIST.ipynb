将Fmnist转换为DBA

```python
from unpickle import unpickle
import matplotlib.pyplot as plt
from PIL import Image 
import numpy as np

path = "D:\code\code_xwd\dataset\Fashion-MNIST\\DBA28x28x1-L\\train"
dict = unpickle(path)
```

```python
type(dict[0])
```

```python
img1 = np.array(dict[0][:])
img2 = np.reshape(img1, (28, 28))
img3 = Image.fromarray(np.uint8(img2),mode="L")

img2[0][0:2] = 255
img2[0][3:5] = 200
img2[2][0:2] = 150
img2[2][3:5] = 100


plt.figure()
plt.imshow(img3)
plt.show()
```
