```python
import numpy as np
import pandas as pd
import os
import csv

data_dir = "C:/Users/Administrator/Desktop"
fname = os.path.join(data_dir, "data.csv")
f=open(fname,encoding='utf-8')
data=pd.read_csv(f)
f.close()
print(data.shape)
print(data.loc[0].values[0])
print("================================================================")
f = open('C:/Users/Administrator/Desktop/result.csv', 'w', newline='', encoding='utf-8')
csv_write = csv.writer(f)

# 定义从正态分布中获取随机数的函数
# 正态分布中的随机数生成,均值-方差
for k in range(0,84,3):
    print(data.loc[k].values[0])
    csv_write.writerow([data.loc[k].values[0],data.loc[k+1].values[0],data.loc[k+1].values[1],data.loc[k+2].values[0],data.loc[k+2].values[1]])
    loc = data.loc[k+1].values[1]
    scale = data.loc[k+2].values[1]
    size=3
    number = np.random.normal(loc=loc, scale=scale, size=size).tolist()
    for i in range(len(number)):
        print(f"{number[i]}")
        csv_write.writerow([number[i]])


f.close()
```
