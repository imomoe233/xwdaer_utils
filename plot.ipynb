```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
filepath = 'D:\\code\\code_xwd\\neurotoxin\\FL_Backdoor_CV\\saved_backdoor_acc_edge_case0_datasetcifar100_save_model_EE3801\\Local_backdoor_cv_cifar100_resnet_snorm0.2_Baseline_Lr0.001_PLr0.003_TLr0.2_AttackNum200_SE1801_AllLayer1.txt'
with open(filepath,'r') as f:
    line = f.readline().lstrip("[")
    nline = line.lstrip("]")
    acc = nline.split(', ')
    acc[1586] = acc[1586].lstrip("]")
for i in range(0, 1587):
    acc[i] = int(float(acc[i]))
```
